```python


student_name = "Ayesha Sharma"

subjects = ["Math", "Physics", "CS", "English", "Chemistry"]
marks = [88, 72, 95, 60, 78]

print("Name:", student_name)
print()

# subjects + grade
for i in range(len(subjects)):
    m = marks[i]

    if m >= 90:
        g = "A+"
    elif m >= 80:
        g = "A"
    elif m >= 70:
        g = "B"
    elif m >= 60:
        g = "C"
    else:
        g = "F"

    print(subjects[i], m, g)

# total & average
total = sum(marks)
avg = total / len(marks)

print("\nTotal:", total)
print("Average:", round(avg, 2))

# highest & lowest
max_m = max(marks)
min_m = min(marks)

print("Highest:", subjects[marks.index(max_m)], max_m)
print("Lowest:", subjects[marks.index(min_m)], min_m)

# add new subjects
count = 0

while True:
    sub = input("\nEnter subject (done to stop): ")
    if sub.lower() == "done":
        break

    m = input("Enter marks: ")

    if not m.isdigit():
        print("Invalid marks")
        continue

    m = int(m)
    if m < 0 or m > 100:
        print("Marks 0-100 only")
        continue

    subjects.append(sub)
    marks.append(m)
    count += 1

print("\nAdded:", count)

new_avg = sum(marks) / len(marks)
print("New Average:", round(new_avg, 2))
```

```python
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]

print("Name              | Average | Status")
print("----------------------------------------")

pass_count = 0
fail_count = 0

top_name = ""
top_avg = 0

total_avg = 0

for data in class_data:
    name = data[0]
    marks = data[1]

    total = 0
    for m in marks:
        total = total + m

    avg = total / len(marks)
    avg = round(avg, 2)

    total_avg = total_avg + avg

    if avg >= 60:
        status = "Pass"
        pass_count = pass_count + 1
    else:
        status = "Fail"
        fail_count = fail_count + 1

    print(name, "|", avg, "|", status)

    if avg > top_avg:
        top_avg = avg
        top_name = name

print()

print("Passed:", pass_count)
print("Failed:", fail_count)

print("Topper:", top_name, "-", top_avg)

class_avg = total_avg / len(class_data)
print("Class Average:", round(class_avg, 2))
```

```python
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "

# step 1 - remove extra spaces
clean_essay = essay.strip()
print("Clean Essay:", clean_essay)

# step 2 - title case
title_text = clean_essay.title()
print("\nTitle Case:", title_text)

# step 3 - count word python
count = clean_essay.count("python")
print("\nCount of 'python':", count)

# step 4 - replace python
new_text = clean_essay.replace("python", "Python 🐍")
print("\nAfter Replace:", new_text)

# step 5 - split into sentences
sentences = clean_essay.split(". ")
print("\nSentences List:", sentences)

# step 6 - print sentences line by line
print("\nSentences:")
i = 1
for s in sentences:
    if not s.endswith("."):
        s = s + "."
    print(i, "-", s)
    i = i + 1
```
